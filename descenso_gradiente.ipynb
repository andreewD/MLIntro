```
import pandas as pd
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

##Creando una funcion anonima 

Funcion anonima:
$F(x,y)=sin(\frac{1}{2}x^2 - \frac{1}{4}y^2)cos(2x+1-e^y)$

```
func = lambda tg: np.sin((1/2)*tg[0]**2- (1/4)*tg[1]**2 +3) * np.cos(2*tg[0]+1 - np.e**tg[1])

sampling = 100

x_1=np.linspace(-2,2,sampling)
x_2=np.linspace(-2,2,sampling)

hyp = np.zeros((sampling,sampling))

for ix1,x1 in enumerate(x_1):
  for ix2,x2 in enumerate(x_2):
    hyp[ix2,ix1]=func([x1,x2])

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X,Y=np.meshgrid(x_1,x_2)
R=np.sqrt(X**2 + Y**2)
ax.plot_surface(X,Y,hyp ,rstride=1,cstride=1,cmap='jet')
```

```
plt.contour(x_1,x_2,hyp,100)
plt.colorbar()
x_0=np.random.rand(2)*4-2
x_0t=np.copy(x_0)
dx = 0.001
lr=0.001
num_interaciones=1000

plt.plot(x_0[0],x_0[1],"o",c='white')
gradiente= np.zeros(2)

for i in range (num_interaciones):
  for it, x0 in enumerate(x_0):
    x_0t = np.copy(x_0)
    x_0[it]=x_0t[it]+dx
    derivada = (func(x_0t)-func(x_0))/dx
    gradiente[it] = derivada

    x_0 = x_0 - lr*gradiente
    
    # print(func(x_0))
    if(i%1000==0):
      plt.plot(x_0[0],x_0[1],"o",c='red')
    
plt.plot(x_0[0],x_0[1],"o",c='blue')
plt.show
```
